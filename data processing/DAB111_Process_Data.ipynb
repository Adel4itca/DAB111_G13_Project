# Data Loading

```python
import pandas as pd
import os
```

```python
# Down load data from  dataset from our GitHub account
url = "https://www.kaggle.com/datasets/jalota/books-dataset?select=Book_Dataset_1.csv"

book_ds = pd.read_csv("Book_Dataset_1.csv")
book_ds.sample(10, random_state=999)
```

# Data Cleaning and Preprocessing
In this section, we describe the data cleaning and preprocessing steps undertaken for this project.

## Data Cleaning Steps<br>
* Drop irrelevant features in our dataset<br>
* Check and rename/ modify some column names<br>
* Check for missing values<br>
* Remove all the rows with missing values If it exists<br>


```python
book_ds.columns
```

After reviewing the data, it became clear to us that these fields are not useful to our project, and we have explained the reason for their deletion.

| Removed Field      | Reason for Removal |
|--------------------|--------------------|
| **Image_Link**     | This column contains external URLs to book cover images. These links provide no analytical value, many are broken or inconsistent, and they are not used anywhere in the project. Therefore, the field was removed to simplify the dataset. |
| **Book_Description** | This field contains long unstructured text, which is not needed for our statistical analysis or visualizations. Processing descriptions requires NLP techniques, which are outside the scope of this project. To improve efficiency and reduce dataset size, the field was removed. |


```python
#Delete unimportant columns
book_ds = book_ds.drop(columns=['Image_Link','Book_Description'])
```

```python
book_ds.sample(5)
```

```python
# Rename columns to shorter names
book_ds = book_ds.rename(columns={
    "Unnamed: 0": "BookID",
    "Title": "Title",
    "Category": "Category",
    "Price": "Price",
    "Price_After_Tax": "PriceTax",
    "Tax_amount": "Tax",
    "Avilability": "Stock",
    "Number_of_reviews": "Reviews",
    "Stars": "Rating"
})

book_ds.sample(5)
```

```python
print(book_ds.dtypes)
```

```python
# Total missing values per column
book_ds.isnull().sum()
```

```python
import os

main_dir = r"D:\ST Clair\DA Courses\DBA 111\DAB111_G13_Project"

data_collection_path = os.path.join(main_dir, "data collection")
os.makedirs(data_collection_path, exist_ok=True)

save_path = os.path.join(data_collection_path, "cl_book_ds.csv")

book_ds.to_csv(save_path, index=False, encoding="utf-8")

print("Saved cleaned dataset to:")
print(save_path)
```
